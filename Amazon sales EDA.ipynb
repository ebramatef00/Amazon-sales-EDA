```python
# Assuming df contains numerical columns
numerical_columns = df.select_dtypes(include='number').columns

# Create subplots
num_plots = len(numerical_columns)
fig, axes = plt.subplots(num_plots, 1, figsize=(8, 6*num_plots))

# Plot each numerical feature
for i, col in enumerate(numerical_columns):
    axes[i].hist(df[col], bins=20, color='skyblue')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Data preprocessing

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

X= df.select_dtypes(include='number')
imp = SimpleImputer(strategy='mean')
X_imputed = imp.fit_transform(X)

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
X= df.select_dtypes(include='object')
print(X.columns)
# Assuming 'df' is your DataFrame with categorical columns
imp_cat = SimpleImputer(strategy='most_frequent')
df_imputed = pd.DataFrame(imp_cat.fit_transform(df), columns=df.columns)
```

```python
df_imputed.isna().sum()
```

```python

# Assuming 'df' contains numerical columns
numerical_columns = df_imputed.select_dtypes(include='number').columns

# Calculate IQR for each numerical column
Q1 = df[numerical_columns].quantile(0.25)
Q3 = df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a mask for outliers
outlier_mask = ((df_imputed[numerical_columns] < lower_bound) | (df_imputed[numerical_columns] > upper_bound))

# Print summary of outliers
print(outlier_mask.sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' contains the relevant columns
# Let's select only the numerical features for this example
numerical_columns = ['Qty', 'Amount']

# Scatter plot (Qty vs. Amount)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Qty', y='Amount', hue='Status', palette='Set2')
plt.xlabel('Quantity')
plt.ylabel('Amount')
plt.title('Scatter Plot: Quantity vs. Amount')
plt.show()

# Line plot for monthly sales trends (assuming 'Date' is the datetime column)
df['Month'] = df['Date'].dt.month
monthly_sales = df.groupby('Month')['Amount'].sum()
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='orange')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.title('Monthly Sales Trends')
plt.grid()
plt.show()
```

```python

# Top-selling categories (assuming 'Category' is the relevant column)
top_categories = df.groupby('Category')['Amount'].sum()
plt.figure(figsize=(8, 6))
top_categories.plot(kind='bar', color='purple')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.title('Selling Categories')
plt.show()
```
